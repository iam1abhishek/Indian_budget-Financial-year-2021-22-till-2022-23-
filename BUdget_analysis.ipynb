```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv(r"E:\yt project\budget analysis\Indian Union Budget FY 21-22 till 23-24.csv")
```

```python
data.head()
```

```python
data.sample(5)
```

```python
data.isnull().sum()
```

```python

# Calculate the mean of the non-null values
Mean_revenue=data['Budget Estimates 2022-2023 Revenue'].mean()
```

```python
data['Budget Estimates 2022-2023 Revenue'].fillna(Mean_revenue,inplace=True)
```

```python
data.isnull().sum()
```

```python
#deleting the null value rows 
data.dropna(subset=['Budget Estimates 2022-2023 Capital' ,'Budget Estimates 2022-2023 Total'],inplace=True)
```

```python
#there  is no any null values 
data.isnull().sum()
```

```python
# Select the top 10 Ministries based on 'Actuals 2021-2022 Revenue'
top_10_ministries = data.nlargest(10, 'Actuals 2021-2022 Revenue\n')

# Plot the bar chart
top_10_ministries.plot.bar(x='Ministry/Department', y='Actuals 2021-2022 Revenue\n', figsize=(10, 6))
plt.xlabel('Ministry/Department')
plt.ylabel('Actuals 2021-2022 Revenue')
plt.title('Top 10 Ministries by Revenue')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Sum up the budget estimates for each ministry
data['Total_Budget'] = data['Budget Estimates 2022-2023 Total']

# Group by 'Ministry/Department' and calculate the total budget estimates
ministry_budget = data.groupby('Ministry/Department')['Total_Budget'].sum()

# Select the top 30 ministries based on total budget estimates
top_10_ministries = ministry_budget.nlargest(10)

# Plot the pie chart
plt.figure(figsize=(10, 8))
plt.pie(top_10_ministries, labels=top_10_ministries.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Budget for Top 10 Ministries')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# Filter the DataFrame to get rows where revenue is 0
ministries_with_zero_revenue = data[data['Actuals 2021-2022 Revenue\n'] == 0]

# Get unique values of 'Ministry/Department' column
distinct_ministries_with_zero_revenue = ministries_with_zero_revenue['Ministry/Department'].unique()

# Print the distinct Ministries/Departments with 0 revenue
print(distinct_ministries_with_zero_revenue)
```

```python
# Select the top 10 Ministries based on 'Actuals 2021-2022 Total'
top_10_ministries = data.nlargest(10, 'Actuals 2021-2022 Total')

# Plot the bar chart
top_10_ministries.plot.bar(x='Ministry/Department', y='Actuals 2021-2022 Total', figsize=(10, 6))
plt.xlabel('Ministry/Department')
plt.ylabel('Actuals 2021-2022 Total')
plt.title('Top 10 Ministries by Revenue')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
# Plot scatter plots between every pair of columns
sns.pairplot(data)
plt.show()
```

```python
# Calculate the correlation matrix
correlation_matrix = data.corr()

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", square=True)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Calculate mean, median, and mode
mean_value = data['Actuals 2021-2022 Total'].mean()
median_value = data['Actuals 2021-2022 Total'].median()
mode_value = data['Actuals 2021-2022 Total'].mode()[0]

# Calculate standard deviation, variance, and range
std_deviation = data['Actuals 2021-2022 Total'].std()
variance = data['Actuals 2021-2022 Total'].var()
range_value = np.ptp(data['Actuals 2021-2022 Total'])  # Peak-to-peak (maximum - minimum)

# Calculate skewness and kurtosis
skewness = data['Actuals 2021-2022 Total'].skew()
kurtosis = data['Actuals 2021-2022 Total'].kurtosis()

# Identify outliers using z-scores (assuming normal distribution)
z_scores = (data['Actuals 2021-2022 Total'] - mean_value) / std_deviation
outliers = data[np.abs(z_scores) > 3]  # Consider values with z-score greater than 3 as outliers

# Print summary statistics
print("Mean:", mean_value)
print("Median:", median_value)
print("Mode:", mode_value)
print("Standard Deviation:", std_deviation)
print("Variance:", variance)
print("Range:", range_value)
print("Skewness:", skewness)
print("Kurtosis:", kurtosis)
print("Number of outliers:", len(outliers))

# Apply logarithmic transformation to the data (if necessary)
data['Actuals 2021-2022 Total'] = data['Actuals 2021-2022 Total'] + 1  # Add a small constant to avoid zero or negative values
log_transformed_data = np.log(data['Actuals 2021-2022 Total'])

# Plot histogram of log-transformed data
plt.figure(figsize=(10, 6))
plt.hist(log_transformed_data, bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Values (Log-transformed)')
plt.ylabel('Frequency')
plt.title('Histogram of Log-transformed Actuals 2021-2022 Total')
plt.grid(True)
plt.show()
```
